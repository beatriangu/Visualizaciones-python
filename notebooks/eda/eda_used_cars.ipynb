# Used Cars Dataset Analysis

This notebook explores the used car dataset through exploratory data analysis, cleaning, and visualization techniques using Python.

```python
df.columns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Price"], bins=30, kde=True)
plt.title("Distribution of Car Prices")
plt.show()
```

```python
df["Price"].describe()
```

```python
df["Brand"].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    y="Brand",
    order=df["Brand"].value_counts().head(10).index
)

plt.title("Top 10 Car Brands")
plt.show()
```

# Price Distribution Analysis
...

```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=df, x="Year", y="Price")

plt.title("Price vs Year")
plt.show()
```

# Price vs Manufacturing Year
...

```python
df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

# Correlation Analysis
...

```python
plt.figure(figsize=(10,5))

sns.boxplot(data=df, x="Body", y="Price")

plt.ylim(0, 80000)

plt.xticks(rotation=45)
plt.title("Price by Car Body Type")
plt.show()
```
