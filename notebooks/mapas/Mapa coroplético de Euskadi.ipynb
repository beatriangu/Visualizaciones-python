```python
import requests
import zipfile
import io

# URL del archivo ZIP
zip_file_url = "https://www.geo.euskadi.eus/cartografia/DatosDescarga/Limites/CB_MUNICIPIOS_5000_ETRS89.zip"
# Get the zipfile from the url
zip_file = requests.get(zip_file_url)
# Descomprimir el archivo zip en la carpeta "/content/"
with zipfile.ZipFile(io.BytesIO(zip_file.content)) as z:
    z.extractall("/content/")
```

```python
import geopandas as gpd

# Ruta al archivo SHP
file_path = "/content/MUNICIPIOS_5000_ETRS89.shp"

# Leer el archivo SHP
shapefile = gpd.read_file(file_path)

# Mostrar información del shapefile
print(shapefile.head())  # Muestra las primeras filas
print(shapefile.crs)     # Sistema de referencia
```

```python
shapefile.plot()
```

```python
import matplotlib.pyplot as plt
# Crear una columna de colores en formato hexadecimal
colores = ('#FF5733,'*259).split(",")[:-1]
colores[0] = "#00FFFF"
colores[15] = "#00FF00"
colores[25] = "#00FF00"

shapefile['color'] = colores  # Ejemplo de colores

# Graficar usando la columna 'color' para colorear los polígonos
shapefile.plot(color=shapefile['color'], edgecolor='black', figsize=(10, 6))

# Mostrar el gráfico
plt.show()
```
